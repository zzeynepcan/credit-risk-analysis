# Exploratory Data Analysis: Credit Risk Analysis

This notebook explores the dataset to identify patterns and relationships between variables, focusing on differences between defaulted and non-defaulted loans.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("Set2")
```

```python
# Load data
df = pd.read_csv('../data/sample_dataset.csv')
df.head()
```

## 1. Credit History vs Default Status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='default_status', y='credit_history', data=df)
plt.title('Credit History by Default Status')
plt.xlabel('Default (0: No, 1: Yes)')
plt.ylabel('Credit History Score')
plt.show()
```

**Observation:** People who default tend to have much lower credit history scores on average compared to non-defaulters.

## 2. Default Rate by Income

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['default_status']==1]['income'], color="red", label="Default", kde=True, stat="density")
sns.histplot(df[df['default_status']==0]['income'], color="blue", label="No Default", kde=True, stat="density")
plt.title('Income Distribution by Default Status')
plt.xlabel('Annual Income')
plt.legend()
plt.show()
```

**Observation:** Defaulting customers are clustered more heavily in the lower income brackets.

```python
# Calculate correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
